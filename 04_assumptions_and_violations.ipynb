### ğŸ’¡ **ì¤‘ìš” ì°¸ê³ ì‚¬í•­**

1. **ê²€ì •ì˜ í•œê³„**: 
   - ì‘ì€ í‘œë³¸: ê²€ì •ë ¥ ë¶€ì¡± (ì´ë¶„ì‚°ì„ ë†“ì¹  ìˆ˜ ìˆìŒ)
   - í° í‘œë³¸: ì‚¬ì†Œí•œ ì°¨ì´ë„ ìœ ì˜í•˜ê²Œ ë‚˜ì˜´

2. **ì‹¤ë¬´ ê¶Œì¥**:
   - ì˜ì‹¬ìŠ¤ëŸ¬ìš°ë©´ Welch's ì‚¬ìš©
   - ë“±ë¶„ì‚° ê²€ì • ê²°ê³¼ì™€ ë¬´ê´€í•˜ê²Œ Welch'sëŠ” ì•ˆì „

````python
# ğŸ”¬ ë“±ë¶„ì‚°ì¼ ë•Œ Student's vs Welch's ë¹„êµ ì‹¤í—˜

import numpy as np
import pandas as pd
from scipy import stats
import plotly.graph_objects as go
from plotly.subplots import make_subplots

def compare_student_vs_welch():
    """ë“±ë¶„ì‚°ì¼ ë•Œ Student'sì™€ Welch's t-test ê²°ê³¼ ë¹„êµ"""
    
    print("=" * 70)
    print("ğŸ”¬ ë“±ë¶„ì‚°ì¼ ë•Œ Student's vs Welch's t-test ë¹„êµ")
    print("=" * 70)
    
    # ì‹œë‚˜ë¦¬ì˜¤ ì„¤ì •
    scenarios = [
        # (n1, n2, mean1, mean2, std, ì„¤ëª…)
        (30, 30, 100, 100, 10, "ë™ì¼ ì¡°ê±´ (Hâ‚€ ì°¸)"),
        (30, 30, 100, 103, 10, "ì‘ì€ ì°¨ì´"),
        (30, 30, 100, 105, 10, "ì¤‘ê°„ ì°¨ì´"),
        (50, 50, 100, 100, 10, "í° í‘œë³¸ (Hâ‚€ ì°¸)"),
        (20, 40, 100, 100, 10, "ë¶ˆê· ë“± í‘œë³¸ (Hâ‚€ ì°¸)"),
        (20, 40, 100, 105, 10, "ë¶ˆê· ë“± í‘œë³¸ + ì°¨ì´")
    ]
    
    results = []
    
    for n1, n2, mean1, mean2, std, description in scenarios:
        # ë“±ë¶„ì‚° ë°ì´í„° ìƒì„± (ê°™ì€ í‘œì¤€í¸ì°¨)
        np.random.seed(42)
        group1 = np.random.normal(mean1, std, n1)
        group2 = np.random.normal(mean2, std, n2)
        
        # ë¶„ì‚° í™•ì¸
        var1 = np.var(group1, ddof=1)
        var2 = np.var(group2, ddof=1)
        var_ratio = max(var1, var2) / min(var1, var2)
        
        # Levene ê²€ì •
        levene_stat, levene_p = stats.levene(group1, group2)
        
        # Student's t-test
        t_student, p_student = stats.ttest_ind(group1, group2, equal_var=True)
        df_student = n1 + n2 - 2
        
        # Welch's t-test
        t_welch, p_welch = stats.ttest_ind(group1, group2, equal_var=False)
        
        # Welch's ììœ ë„ ê³„ì‚°
        s1, s2 = np.std(group1, ddof=1), np.std(group2, ddof=1)
        numerator = (s1**2/n1 + s2**2/n2)**2
        denominator = (s1**2/n1)**2/(n1-1) + (s2**2/n2)**2/(n2-1)
        df_welch = numerator / denominator
        
        results.append({
            'ì‹œë‚˜ë¦¬ì˜¤': description,
            'n1, n2': f"({n1}, {n2})",
            'ë¶„ì‚°ë¹„': f"{var_ratio:.2f}",
            'Levene p': f"{levene_p:.3f}",
            'Student t': f"{t_student:.3f}",
            'Student p': f"{p_student:.4f}",
            'Student df': f"{df_student:.0f}",
            'Welch t': f"{t_welch:.3f}",
            'Welch p': f"{p_welch:.4f}",
            'Welch df': f"{df_welch:.1f}",
            't ì°¨ì´': f"{abs(t_student - t_welch):.4f}",
            'p ì°¨ì´': f"{abs(p_student - p_welch):.4f}"
        })
    
    # DataFrame ìƒì„±
    df_results = pd.DataFrame(results)
    
    print("\nğŸ“Š ê²°ê³¼ í…Œì´ë¸”:")
    print(df_results.to_string(index=False))
    
    # ì‹œê°í™”: 1000íšŒ ì‹œë®¬ë ˆì´ì…˜
    print("\n" + "=" * 70)
    print("ğŸ“ˆ 1000íšŒ ì‹œë®¬ë ˆì´ì…˜: ë“±ë¶„ì‚°ì¼ ë•Œ ë‘ ê²€ì •ì˜ ì¼ì¹˜ë„")
    print("=" * 70)
    
    np.random.seed(42)
    n_simulations = 1000
    t_differences = []
    p_differences = []
    
    for _ in range(n_simulations):
        # ë“±ë¶„ì‚° ë°ì´í„° ìƒì„±
        g1 = np.random.normal(100, 10, 30)
        g2 = np.random.normal(102, 10, 30)  # ê°™ì€ ë¶„ì‚°!
        
        # ë‘ ê²€ì • ìˆ˜í–‰
        t_s, p_s = stats.ttest_ind(g1, g2, equal_var=True)
        t_w, p_w = stats.ttest_ind(g1, g2, equal_var=False)
        
        t_differences.append(abs(t_s - t_w))
        p_differences.append(abs(p_s - p_w))
    
    # íˆìŠ¤í† ê·¸ë¨ ìƒì„±
    fig = make_subplots(
        rows=1, cols=2,
        subplot_titles=[
            't-í†µê³„ëŸ‰ ì°¨ì´ ë¶„í¬',
            'p-ê°’ ì°¨ì´ ë¶„í¬'
        ]
    )
    
    fig.add_trace(
        go.Histogram(x=t_differences, nbinsx=50, 
                     marker_color='blue', opacity=0.7,
                     name='|t_student - t_welch|'),
        row=1, col=1
    )
    
    fig.add_trace(
        go.Histogram(x=p_differences, nbinsx=50,
                     marker_color='green', opacity=0.7,
                     name='|p_student - p_welch|'),
        row=1, col=2
    )
    
    fig.update_layout(
        title='ğŸ”¬ ë“±ë¶„ì‚°ì¼ ë•Œ Student vs Welch ì°¨ì´ (1000íšŒ ì‹œë®¬ë ˆì´ì…˜)',
        height=400
    )
    
    fig.update_xaxes(title_text='t-í†µê³„ëŸ‰ ì ˆëŒ€ ì°¨ì´', row=1, col=1)
    fig.update_xaxes(title_text='p-ê°’ ì ˆëŒ€ ì°¨ì´', row=1, col=2)
    fig.update_yaxes(title_text='ë¹ˆë„', row=1, col=1)
    
    print(f"\nğŸ“Š ì‹œë®¬ë ˆì´ì…˜ ê²°ê³¼:")
    print(f"  â€¢ t-í†µê³„ëŸ‰ í‰ê·  ì°¨ì´: {np.mean(t_differences):.6f}")
    print(f"  â€¢ t-í†µê³„ëŸ‰ ìµœëŒ€ ì°¨ì´: {np.max(t_differences):.6f}")
    print(f"  â€¢ p-ê°’ í‰ê·  ì°¨ì´: {np.mean(p_differences):.6f}")
    print(f"  â€¢ p-ê°’ ìµœëŒ€ ì°¨ì´: {np.max(p_differences):.6f}")
    
    print("\nğŸ’¡ í•µì‹¬ ë°œê²¬:")
    print("  1. ë“±ë¶„ì‚°ì¼ ë•Œ ë‘ ê²€ì •ì˜ t-í†µê³„ëŸ‰ì€ ë™ì¼!")
    print("  2. p-ê°’ë„ ê±°ì˜ ë™ì¼ (ë¯¸ì„¸í•œ ì°¨ì´ëŠ” ììœ ë„ ì°¨ì´ ë•Œë¬¸)")
    print("  3. ê· ë“± í‘œë³¸ì¼ ë•Œ ììœ ë„ë„ ê±°ì˜ ê°™ìŒ")
    print("  4. ê²°ë¡ : ë“±ë¶„ì‚°ì´ë©´ Welch's â‰ˆ Student's")
    print("\nğŸ¯ í•˜ì§€ë§Œ Welch'sëŠ” ì´ë¶„ì‚°ì¼ ë•Œë„ ì•ˆì „í•˜ë¯€ë¡œ")
    print("   â†’ í•­ìƒ Welch's ì‚¬ìš©ì´ ê¶Œì¥ë¨!")
    
    return fig

# ì‹¤í—˜ ì‹¤í–‰
fig = compare_student_vs_welch()
fig.show()

# ====================================================
# ìˆ˜í•™ì  ì¦ëª…: ë“±ë¶„ì‚°ì¼ ë•Œ ì™œ ê°™ì€ê°€?
# ====================================================

print("\n" + "=" * 70)
print("ğŸ“ ìˆ˜í•™ì  ì„¤ëª…: ë“±ë¶„ì‚°ì¼ ë•Œ ì™œ ê²°ê³¼ê°€ ê°™ì€ê°€?")
print("=" * 70)

print("""
Student's t-test:
  t = (XÌ„â‚ - XÌ„â‚‚) / (sp Ã— âˆš(1/nâ‚ + 1/nâ‚‚))
  ì—¬ê¸°ì„œ spÂ² = ((nâ‚-1)sâ‚Â² + (nâ‚‚-1)sâ‚‚Â²) / (nâ‚+nâ‚‚-2)

Welch's t-test:
  t = (XÌ„â‚ - XÌ„â‚‚) / âˆš(sâ‚Â²/nâ‚ + sâ‚‚Â²/nâ‚‚)

ë“±ë¶„ì‚°ì¼ ë•Œ (sâ‚Â² â‰ˆ sâ‚‚Â² = sÂ²):
  
  Student's ë¶„ëª¨:
    sp Ã— âˆš(1/nâ‚ + 1/nâ‚‚) â‰ˆ s Ã— âˆš(1/nâ‚ + 1/nâ‚‚)
  
  Welch's ë¶„ëª¨:
    âˆš(sÂ²/nâ‚ + sÂ²/nâ‚‚) = s Ã— âˆš(1/nâ‚ + 1/nâ‚‚)
  
  â†’ ë‘ ë¶„ëª¨ê°€ ê°™ì•„ì§!
  â†’ t-í†µê³„ëŸ‰ì´ (ê±°ì˜) ë™ì¼!

íŠ¹íˆ nâ‚ = nâ‚‚ì¼ ë•Œ:
  ììœ ë„ë„ ê±°ì˜ ê°™ì•„ì§ (Student: nâ‚+nâ‚‚-2, Welch: â‰ˆ nâ‚+nâ‚‚-2)
  
ê²°ë¡ : ë“±ë¶„ì‚° + ê· ë“±í‘œë³¸ â†’ ë‘ ê²€ì • ì™„ì „ ì¼ì¹˜!
""")
````

## ğŸ›¡ï¸ 2. Welch's t-test: í˜„ì‹¤ì  í•´ë²•

### ğŸ¤” **ë“±ë¶„ì‚°ì„± ìœ„ë°˜ì˜ ë¬¸ì œì **

Student's t-ê²€ì •ì€ ë‘ ê·¸ë£¹ì˜ ë¶„ì‚°ì´ ê°™ë‹¤ê³  ê°€ì •í•©ë‹ˆë‹¤:
$$t = \frac{\bar{X_1} - \bar{X_2}}{s_p \sqrt{\frac{1}{n_1} + \frac{1}{n_2}}}$$

ì—¬ê¸°ì„œ **pooled standard deviation**:
$$s_p = \sqrt{\frac{(n_1-1)s_1^2 + (n_2-1)s_2^2}{n_1 + n_2 - 2}}$$

### âš ï¸ **ë¬¸ì œì **:
- ë‘ ê·¸ë£¹ì˜ ë¶„ì‚°ì´ ì‹¤ì œë¡œ ë‹¤ë¥´ë©´ $s_p$ê°€ ë¶€ì ì ˆ
- ì œ1ì¢… ì˜¤ë¥˜ìœ¨ì´ ëª…ëª©ìˆ˜ì¤€ 0.05ë¥¼ ì´ˆê³¼
- íŠ¹íˆ í‘œë³¸í¬ê¸°ê°€ ë‹¤ë¥¼ ë•Œ ì‹¬ê°

### ğŸš€ **Welchì˜ í•´ë²•**:

$$t = \frac{\bar{X_1} - \bar{X_2}}{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}}$$

**í•µì‹¬ ì•„ì´ë””ì–´**: ê° ê·¸ë£¹ì˜ ë¶„ì‚°ì„ ë”°ë¡œ ê³„ì‚°!

### ğŸ“Š **Satterthwaite ê·¼ì‚¬ë²•**:

ììœ ë„ë¥¼ ë‹¤ìŒê³¼ ê°™ì´ ê·¼ì‚¬:
$$\nu = \frac{\left(\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}\right)^2}{\frac{s_1^4}{n_1^2(n_1-1)} + \frac{s_2^4}{n_2^2(n_2-1)}}$$

**ê²°ê³¼**: ì •ìˆ˜ê°€ ì•„ë‹Œ ììœ ë„ê°€ ë‚˜ì˜¬ ìˆ˜ ìˆìŒ!

````python
# ğŸ”¬ ë“±ë¶„ì‚° vs ì´ë¶„ì‚° ì‹¤ì œ ì˜ˆì œ ì½”ë“œ

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# í•œê¸€ í°íŠ¸ ì„¤ì •
plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['axes.unicode_minus'] = False

# ì‹œë“œ ì„¤ì •
np.random.seed(42)

# ====================================================
# ì˜ˆì œ 1: ëŒ€ê¸°ì—… vs ìŠ¤íƒ€íŠ¸ì—… ì—°ë´‰ ë¹„êµ
# ====================================================

def salary_comparison_example():
    """ëŒ€ê¸°ì—…ê³¼ ìŠ¤íƒ€íŠ¸ì—…ì˜ ì—°ë´‰ ë¶„í¬ ë¹„êµ - ì´ë¶„ì‚°ì˜ ì „í˜•ì  ì˜ˆì œ"""
    
    # ë°ì´í„° ìƒì„±
    # ëŒ€ê¸°ì—…: í‰ê·  6000ë§Œì›, í‘œì¤€í¸ì°¨ 500ë§Œì› (ì²´ê³„ì  ì—°ë´‰ í…Œì´ë¸”)
    corporate_salaries = np.random.normal(6000, 500, 100)
    
    # ìŠ¤íƒ€íŠ¸ì—…: í‰ê·  6200ë§Œì›, í‘œì¤€í¸ì°¨ 1500ë§Œì› (í° ë³€ë™ì„±)
    startup_salaries = np.random.normal(6200, 1500, 80)
    
    # ìŒìˆ˜ ì œê±°
    corporate_salaries = np.abs(corporate_salaries)
    startup_salaries = np.abs(startup_salaries)
    
    # ë“±ë¶„ì‚°ì„± ê²€ì •
    levene_stat, levene_p = stats.levene(corporate_salaries, startup_salaries)
    
    # Student's t-test (ë“±ë¶„ì‚° ê°€ì •)
    t_stat_student, p_val_student = stats.ttest_ind(corporate_salaries, startup_salaries)
    
    # Welch's t-test (ë“±ë¶„ì‚° ê°€ì • ì•ˆí•¨)
    t_stat_welch, p_val_welch = stats.ttest_ind(corporate_salaries, startup_salaries, equal_var=False)
    
    # ì‹œê°í™”
    fig = make_subplots(
        rows=2, cols=2,
        subplot_titles=[
            '1. ì—°ë´‰ ë¶„í¬ ë¹„êµ',
            '2. ë¶„ì‚° ì°¨ì´ ì‹œê°í™”',
            '3. ë°•ìŠ¤í”Œë¡¯ ë¹„êµ',
            '4. ê²€ì • ê²°ê³¼ ë¹„êµ'
        ],
        specs=[
            [{'type': 'histogram'}, {'type': 'scatter'}],
            [{'type': 'box'}, {'type': 'bar'}]
        ]
    )
    
    # 1. íˆìŠ¤í† ê·¸ë¨
    fig.add_trace(
        go.Histogram(x=corporate_salaries, name='ëŒ€ê¸°ì—…', opacity=0.6, 
                     nbinsx=20, marker_color='blue'),
        row=1, col=1
    )
    fig.add_trace(
        go.Histogram(x=startup_salaries, name='ìŠ¤íƒ€íŠ¸ì—…', opacity=0.6,
                     nbinsx=20, marker_color='red'),
        row=1, col=1
    )
    
    # 2. ë¶„ì‚° ë¹„êµ (ì‚°ì ë„)
    fig.add_trace(
        go.Scatter(x=np.random.uniform(0.8, 1.2, len(corporate_salaries)),
                   y=corporate_salaries, mode='markers', name='ëŒ€ê¸°ì—…',
                   marker=dict(color='blue', size=3, opacity=0.5)),
        row=1, col=2
    )
    fig.add_trace(
        go.Scatter(x=np.random.uniform(1.8, 2.2, len(startup_salaries)),
                   y=startup_salaries, mode='markers', name='ìŠ¤íƒ€íŠ¸ì—…',
                   marker=dict(color='red', size=3, opacity=0.5)),
        row=1, col=2
    )
    
    # í‰ê· ì„  ì¶”ê°€
    fig.add_hline(y=np.mean(corporate_salaries), line_dash="solid", 
                  line_color="blue", row=1, col=2)
    fig.add_hline(y=np.mean(startup_salaries), line_dash="solid",
                  line_color="red", row=1, col=2)
    
    # 3. ë°•ìŠ¤í”Œë¡¯
    fig.add_trace(
        go.Box(y=corporate_salaries, name='ëŒ€ê¸°ì—…', marker_color='blue'),
        row=2, col=1
    )
    fig.add_trace(
        go.Box(y=startup_salaries, name='ìŠ¤íƒ€íŠ¸ì—…', marker_color='red'),
        row=2, col=1
    )
    
    # 4. ê²€ì • ê²°ê³¼ ë¹„êµ
    test_names = ['Student\'s t-test', 'Welch\'s t-test']
    p_values = [p_val_student, p_val_welch]
    colors = ['orange' if p > 0.05 else 'green' for p in p_values]
    
    fig.add_trace(
        go.Bar(x=test_names, y=p_values, marker_color=colors,
               text=[f'p={p:.4f}' for p in p_values], textposition='auto'),
        row=2, col=2
    )
    
    # ìœ ì˜ìˆ˜ì¤€ ì„ 
    fig.add_hline(y=0.05, line_dash="dash", line_color="red",
                  annotation_text="Î±=0.05", row=2, col=2)
    
    # ë ˆì´ì•„ì›ƒ ì„¤ì •
    fig.update_layout(
        title='ğŸ¢ ëŒ€ê¸°ì—… vs ìŠ¤íƒ€íŠ¸ì—… ì—°ë´‰ ë¹„êµ (ì´ë¶„ì‚° ì˜ˆì œ)',
        height=800,
        showlegend=True
    )
    
    fig.update_xaxes(title_text='ì—°ë´‰ (ë§Œì›)', row=1, col=1)
    fig.update_xaxes(title_text='ê·¸ë£¹', row=1, col=2)
    fig.update_yaxes(title_text='ì—°ë´‰ (ë§Œì›)', row=1, col=2)
    fig.update_yaxes(title_text='ì—°ë´‰ (ë§Œì›)', row=2, col=1)
    fig.update_yaxes(title_text='p-value', row=2, col=2)
    
    # ê²°ê³¼ ì¶œë ¥
    print("=" * 70)
    print("ğŸ¢ ëŒ€ê¸°ì—… vs ìŠ¤íƒ€íŠ¸ì—… ì—°ë´‰ ë¹„êµ ë¶„ì„")
    print("=" * 70)
    print(f"\nğŸ“Š ê¸°ì´ˆ í†µê³„ëŸ‰:")
    print(f"  ëŒ€ê¸°ì—…: í‰ê·  = {np.mean(corporate_salaries):.0f}ë§Œì›, í‘œì¤€í¸ì°¨ = {np.std(corporate_salaries, ddof=1):.0f}ë§Œì›")
    print(f"  ìŠ¤íƒ€íŠ¸ì—…: í‰ê·  = {np.mean(startup_salaries):.0f}ë§Œì›, í‘œì¤€í¸ì°¨ = {np.std(startup_salaries, ddof=1):.0f}ë§Œì›")
    print(f"\nğŸ” ë¶„ì‚°ë¹„ = {np.var(startup_salaries, ddof=1) / np.var(corporate_salaries, ddof=1):.2f} (ìŠ¤íƒ€íŠ¸ì—…/ëŒ€ê¸°ì—…)")
    print(f"\nğŸ“ˆ Levene ê²€ì • (ë“±ë¶„ì‚°ì„±):")
    print(f"  í†µê³„ëŸ‰ = {levene_stat:.3f}, p-value = {levene_p:.4f}")
    print(f"  ê²°ë¡ : {'ë“±ë¶„ì‚° ê°€ì • ìœ„ë°˜ âŒ' if levene_p < 0.05 else 'ë“±ë¶„ì‚° ê°€ì • ë§Œì¡± âœ…'}")
    print(f"\nğŸ¯ t-ê²€ì • ê²°ê³¼ ë¹„êµ:")
    print(f"  Student's t-test: t = {t_stat_student:.3f}, p = {p_val_student:.4f}")
    print(f"  Welch's t-test:   t = {t_stat_welch:.3f}, p = {p_val_welch:.4f}")
    print(f"\nğŸ’¡ í•´ì„:")
    print(f"  - ë¶„ì‚°ì´ {np.var(startup_salaries, ddof=1) / np.var(corporate_salaries, ddof=1):.1f}ë°° ì°¨ì´ë‚¨")
    print(f"  - ì´ëŸ° ê²½ìš° Welch's t-testê°€ ë” ì ì ˆ")
    print(f"  - {'ë‘ ê²€ì • ëª¨ë‘ ìœ ì˜í•œ ì°¨ì´ ì—†ìŒ' if p_val_welch > 0.05 else 'ìœ ì˜í•œ ì°¨ì´ ìˆìŒ'}")
    
    return fig

# ì˜ˆì œ 1 ì‹¤í–‰
fig1 = salary_comparison_example()
fig1.show()

# ====================================================
# ì˜ˆì œ 2: ì‹ ì•½ vs ìœ„ì•½ ì¹˜ë£Œ íš¨ê³¼ (ê·¹ë‹¨ì  ì´ë¶„ì‚°)
# ====================================================

def drug_effect_example():
    """ì‹ ì•½ê³¼ ìœ„ì•½ì˜ ì¹˜ë£Œ íš¨ê³¼ ë¹„êµ - ê·¹ë‹¨ì  ì´ë¶„ì‚° ì˜ˆì œ"""
    
    # ë°ì´í„° ìƒì„±
    # ìœ„ì•½êµ°: ê±°ì˜ ë³€í™” ì—†ìŒ (í‰ê·  0, í‘œì¤€í¸ì°¨ 2)
    placebo_effect = np.random.normal(0, 2, 50)
    
    # ì‹ ì•½êµ°: ê°œì¸ì°¨ í¼ (í‰ê·  5, í‘œì¤€í¸ì°¨ 10)
    drug_effect = np.random.normal(5, 10, 50)
    
    # í†µê³„ ë¶„ì„
    levene_stat, levene_p = stats.levene(placebo_effect, drug_effect)
    
    # ë‘ ê°€ì§€ t-test
    t_student, p_student = stats.ttest_ind(placebo_effect, drug_effect)
    t_welch, p_welch = stats.ttest_ind(placebo_effect, drug_effect, equal_var=False)
    
    # ììœ ë„ ê³„ì‚°
    n1, n2 = len(placebo_effect), len(drug_effect)
    s1, s2 = np.std(placebo_effect, ddof=1), np.std(drug_effect, ddof=1)
    
    # Student's t-test ììœ ë„
    df_student = n1 + n2 - 2
    
    # Welch's t-test ììœ ë„ (Satterthwaite ê·¼ì‚¬)
    numerator = (s1**2/n1 + s2**2/n2)**2
    denominator = (s1**2/n1)**2/(n1-1) + (s2**2/n2)**2/(n2-1)
    df_welch = numerator / denominator
    
    # ì‹œê°í™”
    fig = make_subplots(
        rows=1, cols=3,
        subplot_titles=[
            'ì¹˜ë£Œ íš¨ê³¼ ë¶„í¬',
            'ë¶„ì‚° ì°¨ì´ (Violin plot)',
            'ììœ ë„ ë¹„êµ'
        ]
    )
    
    # 1. íˆìŠ¤í† ê·¸ë¨
    fig.add_trace(
        go.Histogram(x=placebo_effect, name='ìœ„ì•½', opacity=0.7,
                     nbinsx=20, marker_color='gray'),
        row=1, col=1
    )
    fig.add_trace(
        go.Histogram(x=drug_effect, name='ì‹ ì•½', opacity=0.7,
                     nbinsx=20, marker_color='green'),
        row=1, col=1
    )
    
    # 2. Violin plot
    fig.add_trace(
        go.Violin(y=placebo_effect, name='ìœ„ì•½', box_visible=True,
                  marker_color='gray'),
        row=1, col=2
    )
    fig.add_trace(
        go.Violin(y=drug_effect, name='ì‹ ì•½', box_visible=True,
                  marker_color='green'),
        row=1, col=2
    )
    
    # 3. ììœ ë„ ë¹„êµ
    fig.add_trace(
        go.Bar(x=['Student\'s', 'Welch\'s'], 
               y=[df_student, df_welch],
               text=[f'df={df_student:.0f}', f'df={df_welch:.1f}'],
               textposition='auto',
               marker_color=['blue', 'orange']),
        row=1, col=3
    )
    
    fig.update_layout(
        title='ğŸ’Š ì‹ ì•½ vs ìœ„ì•½ ì¹˜ë£Œ íš¨ê³¼ (ê·¹ë‹¨ì  ì´ë¶„ì‚°)',
        height=400,
        showlegend=True
    )
    
    # ê²°ê³¼ ì¶œë ¥
    print("\n" + "=" * 70)
    print("ğŸ’Š ì‹ ì•½ vs ìœ„ì•½ ì¹˜ë£Œ íš¨ê³¼ ë¶„ì„")
    print("=" * 70)
    print(f"\nğŸ“Š ê¸°ì´ˆ í†µê³„ëŸ‰:")
    print(f"  ìœ„ì•½: í‰ê·  = {np.mean(placebo_effect):.2f}, í‘œì¤€í¸ì°¨ = {s1:.2f}")
    print(f"  ì‹ ì•½: í‰ê·  = {np.mean(drug_effect):.2f}, í‘œì¤€í¸ì°¨ = {s2:.2f}")
    print(f"\nğŸ” ë¶„ì‚°ë¹„ = {s2**2 / s1**2:.2f} (ì‹ ì•½/ìœ„ì•½)")
    print(f"\nğŸ“ˆ ë“±ë¶„ì‚°ì„± ê²€ì •:")
    print(f"  Levene: F = {levene_stat:.3f}, p = {levene_p:.6f}")
    print(f"  ê²°ë¡ : ê·¹ë‹¨ì ì¸ ì´ë¶„ì‚°! âŒâŒâŒ")
    print(f"\nğŸ¯ ììœ ë„ ì°¨ì´:")
    print(f"  Student's t-test: df = {df_student:.0f} (ì •ìˆ˜)")
    print(f"  Welch's t-test:   df = {df_welch:.1f} (ì†Œìˆ˜)")
    print(f"  ì°¨ì´: {df_student - df_welch:.1f} ê°ì†Œ")
    print(f"\nğŸ“Š ê²€ì • ê²°ê³¼:")
    print(f"  Student's: t = {t_student:.3f}, p = {p_student:.4f}")
    print(f"  Welch's:   t = {t_welch:.3f}, p = {p_welch:.4f}")
    print(f"\nâš ï¸ ê²½ê³ : ë¶„ì‚°ë¹„ê°€ {s2**2 / s1**2:.0f}ë°°! Welch's t-test í•„ìˆ˜!")
    
    return fig

# ì˜ˆì œ 2 ì‹¤í–‰
fig2 = drug_effect_example()
fig2.show()

# ====================================================
# ì˜ˆì œ 3: ë“±ë¶„ì‚° vs ì´ë¶„ì‚° ì‹œë®¬ë ˆì´ì…˜ ë¹„êµ
# ====================================================

def variance_simulation():
    """ë“±ë¶„ì‚°ê³¼ ì´ë¶„ì‚° ìƒí™©ì—ì„œì˜ Type I ì˜¤ë¥˜ìœ¨ ë¹„êµ"""
    
    n_simulations = 1000
    alpha = 0.05
    sample_sizes = [(20, 20), (20, 40), (20, 60)]  # ê· ë“±/ë¶ˆê· ë“± í‘œë³¸
    variance_ratios = [1, 2, 4, 8]  # ë¶„ì‚°ë¹„
    
    results = []
    
    for n1, n2 in sample_sizes:
        for var_ratio in variance_ratios:
            student_errors = 0
            welch_errors = 0
            
            for _ in range(n_simulations):
                # H0ê°€ ì°¸ì¸ ìƒí™© (í‰ê· ì€ ê°™ìŒ)
                group1 = np.random.normal(0, 1, n1)
                group2 = np.random.normal(0, np.sqrt(var_ratio), n2)
                
                # Student's t-test
                _, p_student = stats.ttest_ind(group1, group2)
                if p_student < alpha:
                    student_errors += 1
                
                # Welch's t-test
                _, p_welch = stats.ttest_ind(group1, group2, equal_var=False)
                if p_welch < alpha:
                    welch_errors += 1
            
            results.append({
                'Sample Size': f'({n1},{n2})',
                'Variance Ratio': var_ratio,
                'Student Error Rate': student_errors / n_simulations,
                'Welch Error Rate': welch_errors / n_simulations
            })
    
    # ê²°ê³¼ë¥¼ DataFrameìœ¼ë¡œ ë³€í™˜
    df_results = pd.DataFrame(results)
    
    # ì‹œê°í™”
    fig = make_subplots(
        rows=1, cols=3,
        subplot_titles=[
            'ê· ë“± í‘œë³¸ (20,20)',
            'ì¤‘ê°„ ë¶ˆê· ë“± (20,40)', 
            'ê·¹ë‹¨ ë¶ˆê· ë“± (20,60)'
        ]
    )
    
    for i, (n1, n2) in enumerate(sample_sizes):
        sample_data = df_results[df_results['Sample Size'] == f'({n1},{n2})']
        
        fig.add_trace(
            go.Scatter(x=sample_data['Variance Ratio'], 
                      y=sample_data['Student Error Rate'],
                      mode='lines+markers', name=f'Student ({n1},{n2})',
                      line=dict(color='red', dash='solid' if i==0 else 'dash')),
            row=1, col=i+1
        )
        
        fig.add_trace(
            go.Scatter(x=sample_data['Variance Ratio'],
                      y=sample_data['Welch Error Rate'],
                      mode='lines+markers', name=f'Welch ({n1},{n2})',
                      line=dict(color='blue', dash='solid' if i==0 else 'dash')),
            row=1, col=i+1
        )
        
        # ëª…ëª© ì˜¤ë¥˜ìœ¨ ì„ 
        fig.add_hline(y=alpha, line_dash="dash", line_color="gray",
                     annotation_text="Î±=0.05", row=1, col=i+1)
    
    fig.update_layout(
        title='ğŸ”¬ Type I ì˜¤ë¥˜ìœ¨: Student vs Welch (1000íšŒ ì‹œë®¬ë ˆì´ì…˜)',
        height=400
    )
    
    fig.update_xaxes(title_text='ë¶„ì‚°ë¹„', row=1, col=2)
    fig.update_yaxes(title_text='Type I ì˜¤ë¥˜ìœ¨', row=1, col=1)
    
    # ê²°ê³¼ í…Œì´ë¸” ì¶œë ¥
    print("\n" + "=" * 70)
    print("ğŸ”¬ Type I ì˜¤ë¥˜ìœ¨ ì‹œë®¬ë ˆì´ì…˜ ê²°ê³¼")
    print("=" * 70)
    print("\nğŸ“Š ê²°ê³¼ í…Œì´ë¸”:")
    print(df_results.to_string(index=False))
    print("\nğŸ’¡ í•µì‹¬ ë°œê²¬:")
    print("  1. ë“±ë¶„ì‚°(ratio=1)ì¼ ë•Œ: ë‘ ë°©ë²• ëª¨ë‘ Î±=0.05 ìœ ì§€")
    print("  2. ì´ë¶„ì‚° + ê· ë“±í‘œë³¸: Student'sëŠ” ì•½ê°„ ìƒìŠ¹, WelchëŠ” ì•ˆì •")
    print("  3. ì´ë¶„ì‚° + ë¶ˆê· ë“±í‘œë³¸: Student'sëŠ” í¬ê²Œ ìƒìŠ¹, WelchëŠ” ì—¬ì „íˆ ì•ˆì •")
    print("  4. ê²°ë¡ : Welch's t-testê°€ ë” ê°•ê±´í•¨!")
    
    return fig

# ì˜ˆì œ 3 ì‹¤í–‰
fig3 = variance_simulation()
fig3.show()

print("\n" + "ğŸ¯" * 35)
print("\nğŸ“Œ ìµœì¢… ìš”ì•½:")
print("  â€¢ ë“±ë¶„ì‚° ê°€ì •ì´ ìœ„ë°˜ë˜ë©´ Student's t-testëŠ” ë¶€ì •í™•")
print("  â€¢ íŠ¹íˆ í‘œë³¸í¬ê¸°ê¹Œì§€ ë‹¤ë¥´ë©´ ì˜¤ë¥˜ìœ¨ ê¸‰ì¦")
print("  â€¢ Welch's t-testëŠ” ëª¨ë“  ìƒí™©ì—ì„œ ì•ˆì •ì ")
print("  â€¢ ì‹¤ë¬´ ê¶Œì¥: ì˜ì‹¬ìŠ¤ëŸ¬ìš°ë©´ Welch's ì‚¬ìš©!")
print("\n" + "ğŸ¯" * 35)
````

````python
# ====================================================
# ì˜ˆì œ 4: ë¶€ì±„ë„ ë¶„ì„ê³¼ TVLA (Test Vector Leakage Assessment)
# ====================================================

def side_channel_analysis_example():
    """ë¶€ì±„ë„ ë¶„ì„ì—ì„œ Welch's t-testì˜ ì¤‘ìš”ì„±ê³¼ TVLA ì˜ˆì œ"""
    
    print("\n" + "=" * 70)
    print("ğŸ” ë¶€ì±„ë„ ë¶„ì„ê³¼ TVLA (Test Vector Leakage Assessment)")
    print("=" * 70)
    
    # ì•”í˜¸í™” ì¥ì¹˜ì˜ ì „ë ¥ ì†Œë¹„ ì‹œë®¬ë ˆì´ì…˜
    np.random.seed(42)
    
    # ì‹œë‚˜ë¦¬ì˜¤: ë¹„ë°€í‚¤ì˜ íŠ¹ì • ë¹„íŠ¸ê°€ 0ì¼ ë•Œì™€ 1ì¼ ë•Œì˜ ì „ë ¥ ì†Œë¹„
    # ì‹¤ì œ ë¶€ì±„ë„ ê³µê²©ì—ì„œëŠ” ìˆ˜ì²œ~ìˆ˜ë§Œ ê°œì˜ ì¸¡ì •ê°’ì„ ì‚¬ìš©
    
    # í‚¤ ë¹„íŠ¸ = 0ì¼ ë•Œì˜ ì „ë ¥ ì†Œë¹„ (ê¸°ë³¸ ì†Œë¹„ + ì‘ì€ ë…¸ì´ì¦ˆ)
    n_traces_0 = 1000
    power_key0 = np.random.normal(100, 5, n_traces_0)  # í‰ê·  100mW, í‘œì¤€í¸ì°¨ 5mW
    
    # í‚¤ ë¹„íŠ¸ = 1ì¼ ë•Œì˜ ì „ë ¥ ì†Œë¹„ (ì¶”ê°€ ì—°ì‚°ìœ¼ë¡œ ì•½ê°„ ë†’ìŒ + í° ë…¸ì´ì¦ˆ)
    n_traces_1 = 800
    # ì¤‘ìš”: ì‹¤ì œë¡œ ë¶„ì‚°ì´ ë‹¤ë¦„! (ì¶”ê°€ ì—°ì‚°ì‹œ ë³€ë™ì„± ì¦ê°€)
    power_key1 = np.random.normal(102, 15, n_traces_1)  # í‰ê·  102mW, í‘œì¤€í¸ì°¨ 15mW
    
    # ë“±ë¶„ì‚°ì„± ê²€ì •
    levene_stat, levene_p = stats.levene(power_key0, power_key1)
    
    # Student's t-test vs Welch's t-test
    t_student, p_student = stats.ttest_ind(power_key0, power_key1)
    t_welch, p_welch = stats.ttest_ind(power_key0, power_key1, equal_var=False)
    
    # TVLA ê¸°ì¤€: |t| > 4.5 ì´ë©´ ëˆ„ì„¤ íƒì§€
    tvla_threshold = 4.5
    
    print(f"\nğŸ“Š ì „ë ¥ ì†Œë¹„ ë¶„ì„:")
    print(f"  í‚¤=0: í‰ê·  = {np.mean(power_key0):.2f}mW, í‘œì¤€í¸ì°¨ = {np.std(power_key0, ddof=1):.2f}mW")
    print(f"  í‚¤=1: í‰ê·  = {np.mean(power_key1):.2f}mW, í‘œì¤€í¸ì°¨ = {np.std(power_key1, ddof=1):.2f}mW")
    print(f"\nğŸ” ë¶„ì‚°ë¹„ = {np.var(power_key1, ddof=1) / np.var(power_key0, ddof=1):.2f}")
    print(f"  â†’ í‚¤=1ì¼ ë•Œ ë¶„ì‚°ì´ {np.var(power_key1, ddof=1) / np.var(power_key0, ddof=1):.1f}ë°° í¼!")
    
    print(f"\nğŸ“ˆ ë“±ë¶„ì‚°ì„± ê²€ì •:")
    print(f"  Levene: p = {levene_p:.6f}")
    print(f"  ê²°ë¡ : {'ë“±ë¶„ì‚° ê°€ì • ìœ„ë°˜! âŒ' if levene_p < 0.05 else 'ë“±ë¶„ì‚° ê°€ì • ë§Œì¡±'}")
    
    print(f"\nğŸ¯ TVLA ê²€ì • ê²°ê³¼:")
    print(f"  Student's t-test: |t| = {abs(t_student):.3f}")
    print(f"  Welch's t-test:   |t| = {abs(t_welch):.3f}")
    print(f"  TVLA ì„ê³„ê°’: {tvla_threshold}")
    print(f"\n  Student's ê²°ë¡ : {'ëˆ„ì„¤ íƒì§€! ğŸš¨' if abs(t_student) > tvla_threshold else 'ì•ˆì „'}")
    print(f"  Welch's ê²°ë¡ :   {'ëˆ„ì„¤ íƒì§€! ğŸš¨' if abs(t_welch) > tvla_threshold else 'ì•ˆì „'}")
    
    # ì‹œê°í™”
    fig = make_subplots(
        rows=2, cols=2,
        subplot_titles=[
            'ì „ë ¥ ì†Œë¹„ ë¶„í¬',
            'TVLA ê²€ì • ê²°ê³¼',
            'ì‹œê°„ì— ë”°ë¥¸ ì „ë ¥ íŒŒí˜•',
            'ëˆ„ì„¤ íƒì§€ ì„ê³„ê°’'
        ]
    )
    
    # 1. ì „ë ¥ ì†Œë¹„ ë¶„í¬
    fig.add_trace(
        go.Histogram(x=power_key0, name='Key=0', opacity=0.7,
                     nbinsx=30, marker_color='blue'),
        row=1, col=1
    )
    fig.add_trace(
        go.Histogram(x=power_key1, name='Key=1', opacity=0.7,
                     nbinsx=30, marker_color='red'),
        row=1, col=1
    )
    
    # 2. TVLA ê²€ì • ê²°ê³¼ ë¹„êµ
    test_names = ['Student\'s', 'Welch\'s']
    t_values = [abs(t_student), abs(t_welch)]
    colors = ['red' if t > tvla_threshold else 'green' for t in t_values]
    
    fig.add_trace(
        go.Bar(x=test_names, y=t_values, marker_color=colors,
               text=[f'|t|={t:.2f}' for t in t_values], textposition='auto'),
        row=1, col=2
    )
    fig.add_hline(y=tvla_threshold, line_dash="dash", line_color="red",
                  annotation_text="TVLA ì„ê³„ê°’ (4.5)", row=1, col=2)
    
    # 3. ì‹œê°„ì— ë”°ë¥¸ ì „ë ¥ íŒŒí˜• (ì‹¤ì œ ì¸¡ì • ì‹œë®¬ë ˆì´ì…˜)
    time_points = 100
    time = np.arange(time_points)
    
    # í‚¤=0ì¼ ë•Œ í‰ê·  íŒŒí˜•
    waveform_key0 = 100 + 2 * np.sin(time * 0.1) + np.random.normal(0, 0.5, time_points)
    # í‚¤=1ì¼ ë•Œ í‰ê·  íŒŒí˜• (íŠ¹ì • ì‹œì ì— ìŠ¤íŒŒì´í¬)
    waveform_key1 = 100 + 2 * np.sin(time * 0.1) + np.random.normal(0, 0.5, time_points)
    waveform_key1[40:45] += 5  # ì•”í˜¸ ì—°ì‚° ì‹œì 
    
    fig.add_trace(
        go.Scatter(x=time, y=waveform_key0, mode='lines', name='Key=0',
                   line=dict(color='blue', width=2)),
        row=2, col=1
    )
    fig.add_trace(
        go.Scatter(x=time, y=waveform_key1, mode='lines', name='Key=1',
                   line=dict(color='red', width=2)),
        row=2, col=1
    )
    
    # 4. ë‹¤ì–‘í•œ ë…¸ì´ì¦ˆ ë ˆë²¨ì—ì„œì˜ íƒì§€ ì„±ëŠ¥
    noise_levels = np.arange(1, 20, 2)
    detection_student = []
    detection_welch = []
    
    for noise in noise_levels:
        # 100íšŒ ë°˜ë³µí•˜ì—¬ í‰ê·  íƒì§€ìœ¨ ê³„ì‚°
        student_detect = 0
        welch_detect = 0
        
        for _ in range(100):
            sample0 = np.random.normal(100, 5, 100)
            sample1 = np.random.normal(102, noise, 100)
            
            t_s, _ = stats.ttest_ind(sample0, sample1)
            t_w, _ = stats.ttest_ind(sample0, sample1, equal_var=False)
            
            if abs(t_s) > tvla_threshold:
                student_detect += 1
            if abs(t_w) > tvla_threshold:
                welch_detect += 1
        
        detection_student.append(student_detect)
        detection_welch.append(welch_detect)
    
    fig.add_trace(
        go.Scatter(x=noise_levels, y=detection_student, mode='lines+markers',
                   name='Student\'s', line=dict(color='orange', width=2)),
        row=2, col=2
    )
    fig.add_trace(
        go.Scatter(x=noise_levels, y=detection_welch, mode='lines+markers',
                   name='Welch\'s', line=dict(color='green', width=2)),
        row=2, col=2
    )
    
    # ë ˆì´ì•„ì›ƒ ì„¤ì •
    fig.update_layout(
        title='ğŸ” ë¶€ì±„ë„ ë¶„ì„: TVLAë¥¼ ì´ìš©í•œ ì •ë³´ ëˆ„ì„¤ íƒì§€',
        height=800
    )
    
    fig.update_xaxes(title_text='ì „ë ¥ ì†Œë¹„ (mW)', row=1, col=1)
    fig.update_yaxes(title_text='|t-statistic|', row=1, col=2)
    fig.update_xaxes(title_text='ì‹œê°„ (ìƒ˜í”Œ)', row=2, col=1)
    fig.update_yaxes(title_text='ì „ë ¥ (mW)', row=2, col=1)
    fig.update_xaxes(title_text='ë…¸ì´ì¦ˆ ë ˆë²¨ (Ïƒ)', row=2, col=2)
    fig.update_yaxes(title_text='íƒì§€ íšŸìˆ˜ (/100)', row=2, col=2)
    
    print("\nğŸ’¡ ë¶€ì±„ë„ ë¶„ì„ì—ì„œ Welch's t-testê°€ ì¤‘ìš”í•œ ì´ìœ :")
    print("  1. ì•”í˜¸ ì—°ì‚°ì˜ ë³µì¡ë„ì— ë”°ë¼ ì „ë ¥ ì†Œë¹„ ë¶„ì‚°ì´ ë‹¤ë¦„")
    print("  2. ì¸¡ì • í™˜ê²½ ë…¸ì´ì¦ˆê°€ ìƒí™©ì— ë”°ë¼ ë³€í•¨")
    print("  3. ì˜ëª»ëœ í†µê³„ ê²€ì •ì€ ë³´ì•ˆ ì·¨ì•½ì ì„ ë†“ì¹  ìˆ˜ ìˆìŒ")
    print("  4. TVLA í‘œì¤€ì—ì„œë„ Welch's t-test ê¶Œì¥")
    
    return fig

# ì˜ˆì œ 4 ì‹¤í–‰
fig4 = side_channel_analysis_example()
fig4.show()

# ====================================================
# ì˜ˆì œ 5: ê³ ê¸‰ TVLA - 1st Order vs Higher Order ë¶„ì„
# ====================================================

def advanced_tvla_example():
    """ê³ ê¸‰ TVLA: 1ì°¨ ë° ê³ ì°¨ ë¶€ì±„ë„ ë¶„ì„"""
    
    print("\n" + "=" * 70)
    print("ğŸ”¬ ê³ ê¸‰ TVLA: ë§ˆìŠ¤í‚¹ ëŒ€ì‘ì±… í‰ê°€")
    print("=" * 70)
    
    np.random.seed(42)
    n_traces = 1000
    
    # ì‹œë‚˜ë¦¬ì˜¤: ë§ˆìŠ¤í‚¹ëœ ì•”í˜¸ êµ¬í˜„ í‰ê°€
    # 1ì°¨ ë§ˆìŠ¤í‚¹ì€ 1st order ê³µê²©ì„ ë°©ì–´í•˜ì§€ë§Œ 2nd orderì—ëŠ” ì·¨ì•½
    
    # ê³ ì • vs ëœë¤ ë°ì´í„°ì—ì„œì˜ ì „ë ¥ ì†Œë¹„
    # ë§ˆìŠ¤í‚¹ ì—†ìŒ (ì·¨ì•½í•œ êµ¬í˜„)
    unmasked_fixed = np.random.normal(100, 3, n_traces)
    unmasked_random = np.random.normal(105, 3, n_traces)  # ëª…í™•í•œ ì°¨ì´
    
    # 1ì°¨ ë§ˆìŠ¤í‚¹ (1st order ë°©ì–´)
    masked_fixed = np.random.normal(100, 5, n_traces)
    masked_random = np.random.normal(100.5, 5, n_traces)  # ì‘ì€ ì°¨ì´
    
    # 2nd order ë¶„ì„ì„ ìœ„í•œ ì œê³±ê°’
    masked_fixed_sq = (masked_fixed - np.mean(masked_fixed))**2
    masked_random_sq = (masked_random - np.mean(masked_random))**2
    # ì œê³±í•˜ë©´ ë¶„ì‚°ì´ ë‹¬ë¼ì§!
    masked_random_sq = masked_random_sq * 1.5  # ì¸ìœ„ì ìœ¼ë¡œ ì°¨ì´ ìƒì„±
    
    # 1st order TVLA
    t_unmasked_student, _ = stats.ttest_ind(unmasked_fixed, unmasked_random)
    t_unmasked_welch, _ = stats.ttest_ind(unmasked_fixed, unmasked_random, equal_var=False)
    
    t_masked_student, _ = stats.ttest_ind(masked_fixed, masked_random)
    t_masked_welch, _ = stats.ttest_ind(masked_fixed, masked_random, equal_var=False)
    
    # 2nd order TVLA (ì œê³±ê°’ì— ëŒ€í•œ ë¶„ì„)
    t_2nd_student, _ = stats.ttest_ind(masked_fixed_sq, masked_random_sq)
    t_2nd_welch, _ = stats.ttest_ind(masked_fixed_sq, masked_random_sq, equal_var=False)
    
    # Levene ê²€ì •
    _, levene_2nd = stats.levene(masked_fixed_sq, masked_random_sq)
    
    print(f"\nğŸ“Š 1st Order TVLA ê²°ê³¼:")
    print(f"  ë§ˆìŠ¤í‚¹ ì—†ìŒ:")
    print(f"    Student's: |t| = {abs(t_unmasked_student):.2f} {'ğŸš¨ ëˆ„ì„¤!' if abs(t_unmasked_student) > 4.5 else 'âœ…'}")
    print(f"    Welch's:   |t| = {abs(t_unmasked_welch):.2f} {'ğŸš¨ ëˆ„ì„¤!' if abs(t_unmasked_welch) > 4.5 else 'âœ…'}")
    print(f"  ë§ˆìŠ¤í‚¹ ì ìš©:")
    print(f"    Student's: |t| = {abs(t_masked_student):.2f} {'ğŸš¨ ëˆ„ì„¤!' if abs(t_masked_student) > 4.5 else 'âœ… ì•ˆì „'}")
    print(f"    Welch's:   |t| = {abs(t_masked_welch):.2f} {'ğŸš¨ ëˆ„ì„¤!' if abs(t_masked_welch) > 4.5 else 'âœ… ì•ˆì „'}")
    
    print(f"\nğŸ“Š 2nd Order TVLA ê²°ê³¼ (ì œê³±ê°’ ë¶„ì„):")
    print(f"  ë¶„ì‚°ë¹„: {np.var(masked_random_sq) / np.var(masked_fixed_sq):.2f}")
    print(f"  Levene p-value: {levene_2nd:.6f} {'â†’ ì´ë¶„ì‚°!' if levene_2nd < 0.05 else ''}")
    print(f"  Student's: |t| = {abs(t_2nd_student):.2f} {'ğŸš¨ 2ì°¨ ëˆ„ì„¤!' if abs(t_2nd_student) > 4.5 else 'âœ…'}")
    print(f"  Welch's:   |t| = {abs(t_2nd_welch):.2f} {'ğŸš¨ 2ì°¨ ëˆ„ì„¤!' if abs(t_2nd_welch) > 4.5 else 'âœ…'}")
    
    # ì‹œê°í™”
    fig = make_subplots(
        rows=2, cols=2,
        subplot_titles=[
            'ë§ˆìŠ¤í‚¹ ì—†ìŒ: 1st Order',
            'ë§ˆìŠ¤í‚¹ ì ìš©: 1st Order',
            'ë§ˆìŠ¤í‚¹ ì ìš©: 2nd Order (ì›ë³¸)',
            'ë§ˆìŠ¤í‚¹ ì ìš©: 2nd Order (ì œê³±)'
        ]
    )
    
    # 1. ë§ˆìŠ¤í‚¹ ì—†ìŒ
    fig.add_trace(
        go.Histogram(x=unmasked_fixed, name='Fixed', opacity=0.6,
                     marker_color='blue', nbinsx=30),
        row=1, col=1
    )
    fig.add_trace(
        go.Histogram(x=unmasked_random, name='Random', opacity=0.6,
                     marker_color='red', nbinsx=30),
        row=1, col=1
    )
    
    # 2. 1ì°¨ ë§ˆìŠ¤í‚¹ (1st order)
    fig.add_trace(
        go.Histogram(x=masked_fixed, name='Fixed', opacity=0.6,
                     marker_color='blue', nbinsx=30),
        row=1, col=2
    )
    fig.add_trace(
        go.Histogram(x=masked_random, name='Random', opacity=0.6,
                     marker_color='red', nbinsx=30),
        row=1, col=2
    )
    
    # 3. 1ì°¨ ë§ˆìŠ¤í‚¹ ì›ë³¸ ë°ì´í„°
    fig.add_trace(
        go.Box(y=masked_fixed, name='Fixed', marker_color='blue'),
        row=2, col=1
    )
    fig.add_trace(
        go.Box(y=masked_random, name='Random', marker_color='red'),
        row=2, col=1
    )
    
    # 4. 2nd order (ì œê³±ê°’)
    fig.add_trace(
        go.Box(y=masked_fixed_sq, name='FixedÂ²', marker_color='blue'),
        row=2, col=2
    )
    fig.add_trace(
        go.Box(y=masked_random_sq, name='RandomÂ²', marker_color='red'),
        row=2, col=2
    )
    
    fig.update_layout(
        title='ğŸ”¬ ê³ ê¸‰ TVLA: 1st vs 2nd Order ë¶€ì±„ë„ ë¶„ì„',
        height=700
    )
    
    print("\nğŸ’¡ í•µì‹¬ í†µì°°:")
    print("  1. 2nd order ë¶„ì„ì—ì„œëŠ” ë¹„ì„ í˜• ë³€í™˜ìœ¼ë¡œ ë¶„ì‚°ì´ í¬ê²Œ ë‹¬ë¼ì§")
    print("  2. Student's t-testëŠ” ì´ë¶„ì‚° ìƒí™©ì—ì„œ ë¶€ì •í™•í•  ìˆ˜ ìˆìŒ")
    print("  3. Welch's t-testê°€ ë” ì‹ ë¢°í•  ìˆ˜ ìˆëŠ” ê²°ê³¼ ì œê³µ")
    print("  4. ë³´ì•ˆ í‰ê°€ì—ì„œ í†µê³„ì  ì •í™•ì„±ì€ ë§¤ìš° ì¤‘ìš”!")
    
    return fig

# ì˜ˆì œ 5 ì‹¤í–‰
fig5 = advanced_tvla_example()
fig5.show()

print("\n" + "ğŸ”" * 35)
print("\nğŸ“Œ ë¶€ì±„ë„ ë¶„ì„ê³¼ TVLA ìµœì¢… ìš”ì•½:")
print("  â€¢ TVLAëŠ” ì•”í˜¸ êµ¬í˜„ì˜ ë¶€ì±„ë„ ëˆ„ì„¤ì„ íƒì§€í•˜ëŠ” í‘œì¤€ ë°©ë²•")
print("  â€¢ |t| > 4.5ë¥¼ ì„ê³„ê°’ìœ¼ë¡œ ì‚¬ìš© (99.999% ì‹ ë¢°ìˆ˜ì¤€)")
print("  â€¢ ì•”í˜¸ ì—°ì‚°ì€ ë³¸ì§ˆì ìœ¼ë¡œ ì´ë¶„ì‚°ì„ ìƒì„±")
print("  â€¢ Welch's t-testê°€ ë” ì •í™•í•œ ë³´ì•ˆ í‰ê°€ ì œê³µ")
print("  â€¢ ì˜ëª»ëœ í†µê³„ëŠ” ë³´ì•ˆ ì·¨ì•½ì ì„ ë†“ì¹  ìœ„í—˜!")
print("\n" + "ğŸ”" * 35)
````
